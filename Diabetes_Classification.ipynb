## **Data Splitting (Train/Test Datasets)**

```python
# isolating target variable which is Diabetes
x = df.drop('Diabetes', axis=1)
y = df.get('Diabetes')
```

```python
# creating training and testing datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

## **Modeling: Decision Trees**

```python
from sklearn.tree import DecisionTreeClassifier

# creating and fitting Decision Trees for my classification algorithm
diabetesTree = DecisionTreeClassifier()

diabetesTree.fit(X_train, y_train)
```

```python
from sklearn import metrics

# creating predictions for training and testing data
training_v1 = diabetesTree.predict(X_train)
pred_v1 = diabetesTree.predict(X_test)

# getting accuracies for training and testing
print("Training Accuracy:", metrics.accuracy_score(y_train, training_v1))
print("Testing Accuracy:", metrics.accuracy_score(y_test, pred_v1))
```

- There seems to be signs of overfitting so I will perform hyperparameter tuning to address this issue

## **Hyperparameter Tuning**

```python
# performing hyperparameter tuning through Grid Search
from sklearn.model_selection import GridSearchCV

# selecting Decision Tree and labeling various hyperparameters
decisionTree = DecisionTreeClassifier()
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': [2, 4, 6, 8, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'max_features': [None, 'sqrt', 'log2'],
}

# creating grid search object
grid_search = GridSearchCV(estimator=decisionTree, param_grid=param_grid, cv=3, scoring='accuracy')

# fitting grid search model with the data
grid_search.fit(X_train,y_train)
```

```python
# get optimized model
optimized_model = grid_search.best_estimator_

# find best parameters for that model
parameters = grid_search.best_params_
print('Best parameters:', parameters)
```

```python
# create predictions
training_v2 = optimized_model.predict(X_train)
pred_v2 = optimized_model.predict(X_test)

# get accuracy on new model
print("Training Accuracy:", metrics.accuracy_score(y_train, training_v2))
print("Testing Accuracy:", metrics.accuracy_score(y_test, pred_v2))
```
